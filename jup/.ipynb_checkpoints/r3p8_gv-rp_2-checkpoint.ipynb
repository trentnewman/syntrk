```python
import os
import subprocess
import numpy as np
import pandas as pd
import re

from Bio import SeqIO
from pygenomeviz import GenomeViz

from matplotlib.pyplot import figure
import matplotlib.pyplot as plt
import matplotlib.patheffects as pe
import matplotlib.patches as patches
```

```python
plt.rc('text', usetex=True)
plt.rc(
    'text.latex', 
    preamble=r'''
    \usepackage{dejavu}
    \renewcommand{\familydefault}{\sfdefault}
    \usepackage{mathastext}
    ''')
```

```python
tst = 1
tstt = 'test'+str(tst)
```

```python
regs = pd.read_csv('../extseq/'+str(tst)+'/regions.csv')
regs = regs.drop(columns=['st', 'en'])
regs.rename(columns={'est':'st'}, inplace=True)
regs.rename(columns={'een':'en'}, inplace=True)
regs['lennie'] = regs['en']-regs['st']
```

```python
links = pd.read_csv('../'+tstt+'/coords_foc.tsv', sep='\t')
links = links[links['REF_NAME']!=links['QUERY_NAME']]
links = links.drop(columns=['index'])
links = links.reset_index(drop=True)
links = links.reset_index(drop=False)
links.tail()
```

```python
rrr = []
rs = []
re = []
# ri = []
qqq = []
qs = []
qe = []
# qi = []
for l in range(len(links['index'])):
    #
    ref = links.iloc[l]['REF_NAME']
    ref = ref.split('.fasta')[0]
    ref = ref.replace('_foc','')
    refs = regs[regs['nm']==ref]['st'].iloc[0]
#     refs = refs - refs
    refe = regs[regs['nm']==ref]['en'].iloc[0]
#     refe = refe - refs
    rrr.append(ref)
    rs.append(refs)
    re.append(refe)
    #
    qur = links.iloc[l]['QUERY_NAME']
    qur = qur.split('.fasta')[0]
    qur = qur.replace('_foc','')
    qurs = regs[regs['nm']==qur]['st'].iloc[0]
#     qurs = qurs - qurs
    qure = regs[regs['nm']==qur]['en'].iloc[0]
#     qure = qure - qurs
    qqq.append(qur)
    qs.append(qurs)
    qe.append(qure)
#
links['rrr'] = rrr
links['rs'] = rs
links['re'] = re
links['qqq'] = qqq
links['qs'] = qs
links['qe'] = qe
links.head(2)
#
slinks = links
```

```python
annie = pd.read_csv('../ann/ann_comb'+str(tst)+'_region.csv', sep='\t')
asn = []
aen = []
for i in range(len(annie['start'])):
    nstt = regs[regs['nm']==annie.iloc[i]['sp']]['st'].iloc[0]
    asn.append(annie.iloc[i]['start_rc']-nstt)
    aen.append(annie.iloc[i]['end_rc']-nstt)
annie['startn'] = asn
annie['endn'] = aen
```

```python
ganni = pd.read_csv('../ann/ann_comb'+str(tst)+'_PRKACB.csv', sep='\t')
ganni['gee'] = 'PRKACB'
lanni = pd.read_csv('../ann/ann_comb'+str(tst)+'_lncPRKACB.csv', sep='\t')
lanni['gee'] = 'lncPRKACB'
panni = pd.concat([ganni,lanni])
pstn = []
penn = []
for p in range(len(panni['sp'])):
    rgst = regs[regs['nm']==panni.iloc[p]['sp']].iloc[0]['st']
    rgen = regs[regs['nm']==panni.iloc[p]['sp']].iloc[0]['en']
    if panni.iloc[p]['dir'] == 'f':
        pstn.append(panni.iloc[p]['start_rc']-rgst)
        penn.append(panni.iloc[p]['end_rc']-rgst)
    if panni.iloc[p]['dir'] == 'r':
#         pstn.append(panni.iloc[p]['start_rc']-rgst)
#         penn.append(panni.iloc[p]['end_rc']-rgst)
        pstn.append(rgen-panni.iloc[p]['end'])
        penn.append(rgen-panni.iloc[p]['start'])
panni['startn']=pstn
panni['endn']=penn
spanni = panni
spanni = spanni[spanni['sp']=='Tammar']
spanni
```

```python
# rdir = '../rpout/'+str(tst)+'/'
rdir = '../repeats/'+str(tst)+'/mammals/'
rpfls = os.listdir(rdir)
rpfls = sorted(rpfls)
rpcmb = pd.read_csv(rdir+'rm_spec_r_comb.csv', sep='\t')
rpcmb['strand'] = rpcmb['match'].replace('C','-1').replace('+','+1')
rpcmb['strand'] = rpcmb['strand'].astype(int)
# rpcmb
```

```python
tdir = '../tandem/'+str(tst)+'/'
tfl = os.listdir(tdir)
tdf = pd.read_csv(tdir+'trf_cmb.csv', sep='\t')
# tdf
```

```python
gv = GenomeViz(    
    fig_width=15.038, # brd has figw 10 at 0.665\textwidth 
    feature_track_ratio=1, # keeping constant 
    fig_track_height=0.19, #0.17
    link_track_ratio=0.8, #0.65 
    tick_style="axis",
    align_type='center',
#     track_spines=True
    tick_labelsize = 17,
)

uss = regs['nm'].unique()
# uss = ['Human','Mouse','Elephant'] # subset
# uss = ['Human','Mouse'] # subset
# uss = ['Human'] # subset

sregs = regs
sregs = sregs.reset_index(drop=False)
sregs = sregs[sregs['nm'].isin(uss)] 
for i in range(len(sregs['index'])):
    ssregs = sregs.iloc[i]
    track1 = gv.add_feature_track(
#         name=ssregs['nm']+'1', 
        name=ssregs['nm'], 
        size=ssregs['lennie'],
        linewidth=0,
#         labelsize=0
    )
    # Custom track label position
    track1.labelsize=0 # Hide main label
    track1.set_sublabel(
        ssregs['nm'],
        sublabel_kws=dict(
            alpha=1, y=-2, 
            x=-1000, # position of track label away from track
            va='center', ha='right',
            fontsize=22
        )
    )
    
#     track1.add_subtrack(name=ssregs['nm']) # Add subtrack
    track2 = gv.add_feature_track(
#         name=sglist['ff'], size=sglist['ss']
        name=ssregs['nm']+'2', size=ssregs['lennie'],
        linewidth=0,
        link_track_ratio=0,
        labelsize=0
    )
    track3 = gv.add_feature_track(
#         name=sglist['ff'], size=sglist['ss']
        name=ssregs['nm']+'3', size=ssregs['lennie'],
        linewidth=0,
        link_track_ratio=0,
        labelsize=0
    )
    track3.add_subtrack(name=ssregs['nm']) # Add subtrack
    
    # if ssregs['nm'] == 'Human':
    #     track1.set_sublabel(
    #         'Annotation',
    #         sublabel_kws=dict(
    #             alpha=1, y=0.5, x=200, va='top', ha='left') # zorder didn't work
    #     )
    #     track2.set_sublabel(
    #         'Transposons',
    #         sublabel_kws=dict(
    #             alpha=1, y=0.5, x=-200, va='center', ha='right') # zorder didn't work
    #     )
    #     track3.set_sublabel(
    #         'Tandem repeats',
    #         sublabel_kws=dict(
    #             alpha=1,  y=0.5, x=-200, va='center', ha='right')
    #     )
        
    track1.ratio=1
    track2.ratio=0.5
    track3.ratio=0.5
                 
    # ANN  
    spanni = panni
    spanni = spanni[spanni['sp']==ssregs['nm']]   
    spanni = spanni.reset_index(drop=True)
    spanni = spanni.reset_index(drop=False) 
    for ge in range(len(spanni['gee'].unique())):
        gee = spanni['gee'].unique()[ge]
        sgpanni = spanni[spanni['gee']==gee]
        for an in range(len(sgpanni['index'])):
            sg = sgpanni.iloc[an]
            if sg['featuretype'] == 'unk_exA':
                if sg['startn'] >= 0:
                    if sg['startn'] < ssregs['lennie']:
                        if sg['endn'] <= ssregs['lennie']:
                            track1.add_feature(
                                sg['startn']-1, sg['endn']-1,
                                strand=sg['pm'], 
                                plotstyle='bigarrow', 
                                label='', 
                                arrow_shaft_ratio=1.0, size_ratio=0.75,
                                facecolor='grey', linewidth=2,
                                patch_kws=dict(zorder=1, joinstyle='round')
                            )
                        elif sg['endn'] > ssregs['lennie']:
                            print('edit annot code')
            elif sg['featuretype'] == 'unk_inA':
                if sg['startn'] >= 0:
                    if sg['startn'] < ssregs['lennie']:
                        if sg['endn'] <= ssregs['lennie']:
                            track1.add_feature(
                                sg['startn']-1, sg['endn']-1,
                                strand=sg['pm'], 
                                plotstyle='bigarrow', 
                                label='', 
                                arrow_shaft_ratio=1.0, size_ratio=0.4,
                                facecolor='lightgrey', linewidth=0,
                                patch_kws=dict(zorder=0, joinstyle='round')
                            )
                        elif sg['endn'] > ssregs['lennie']:
                            track1.add_feature(
                                sg['startn']-1, ssregs['lennie'],
                                strand=sg['pm'], 
                                plotstyle='bigbox', 
                                label='', 
                                arrow_shaft_ratio=1.0, size_ratio=0.4,
                                facecolor='lightgrey', linewidth=0,
                                patch_kws=dict(zorder=0, joinstyle='round')
                            )
    # RM
    rpn = rpcmb[rpcmb['spec']==ssregs['nm']]
    rpn = rpn.reset_index(drop=True)
    rpn = rpn.reset_index(drop=False)
    for an in range(len(rpn['index'])):
        sg = rpn.iloc[an]
        if sg['qpos_begin'] >= 0:
            if sg['qpos_begin'] < ssregs['lennie']:
                if sg['rep_group'] == 'LTR':
                    fcoly = 'magenta'
                elif sg['rep_group'] == 'LINE/L2':
                    fcoly = 'cyan'
                elif sg['rep_group'] == 'Oth. LINE':
                    fcoly = 'blue'
                elif sg['rep_group'] == 'SINE':
                    fcoly = 'darkgreen'
                elif sg['rep_group'] == 'Other\nrepeat':
                    fcoly = 'gold'
                elif sg['rep_group'] == 'Prob.':
                    fcoly = 'grey'
                else:
                    fcoly = 'white'
                # repeatmasker repeats
                track2.add_feature(
                    sg['qpos_begin']-1, sg['qpos_end']-1, 
                    strand=sg['strand'], 
                    label='', labelrotation=0, 
                    labelvpos='bottom', labelha='center',
                    arrow_shaft_ratio=1.0, size_ratio=0.9,
                    facecolor=fcoly, linewidth=0.5,
                    patch_kws=dict(zorder=1, alpha=0.5, joinstyle='round')
                )
    # TRF   
    stdf = tdf[tdf['f']==ssregs['nm']+'_foc']   
    stdf = stdf.reset_index(drop=True)
    stdf = stdf.reset_index(drop=False)   
    for an in range(len(stdf['index'])):
        sg = stdf.iloc[an]
        if sg['i_st'] >= 0:
            if sg['i_st'] < ssregs['lennie']:
                # tandem repeat finder repeats
                track3.add_feature(
                    sg['i_st']-1, sg['i_en']-1, 
                    label='', labelrotation=0, 
                    labelvpos='bottom', labelha='center',
                    plotstyle='bigbox', facecolor='grey', linewidth=0,
#                     size_ratio=0.5,
                    patch_kws=dict(zorder=1, alpha=0.45)
                )
                
# Add links
sslinks = slinks[ (slinks['rrr'].isin(uss)) & (slinks['qqq'].isin(uss)) ]
for i in range(len(sslinks['index'])):
    sslink = sslinks.iloc[i]
    link_data1 = (sslink['rrr']+'3', sslink['REF_START'], sslink['REF_END']-1)
    link_data2 = (sslink['qqq'], sslink['QUERY_START'], sslink['QUERY_END']-1)
    gv.add_link(link_data1, link_data2, 
                normal_color="#2A2A2A", inverted_color="orange", 
                curve=True)  
#    
# position of plot fig is important
fig = gv.plotfig(dpi=300)
#
#

# Custom subtrack additions
for s in range(len(sregs['nm'])):
    snn = sregs['nm'][s]
    for track in gv.get_feature_tracks():
#         print(track.name)
        if track.name == snn+'3':
            subtrack = track.get_subtrack(sregs['nm'][s])
            x = np.arange(0, track.end, 100)
            y = np.random.randint(0, 100, len(x))
            # CpG density
            cf = pd.read_csv('../extseq/'+str(tst)+'/'+snn+'_foc_cpg1.csv')
#             subtrack.ax.fill_between(x, y, color="grey")
            subtrack.ax.fill_between(
#                 cf['xi'], # when left align
                subtrack.offset+cf['xi'], 
                cf['yi'], color="white")
            subtrack.ax.plot(
                subtrack.offset+cf['xi'], 
                cf['yi'], color="black", linewidth=2)
#             subtrack.ax.spines["bottom"].set_visible(True)
            subtrack.ax.axhline(y=0, xmin=0, xmax=100, 
                                c='black', lw=3)
            # if sregs['nm'][s] == 'Human':
            #     subtrack.ax.text(
            #         x=200, y=10, s='CpG density', 
            #         va='center', ha='left',size=15)
            subtrack.ax.set_ylim(0,cf['yi'].max())
            # Track separators
            tx_1 = -300
            tx_2 = -400
            t_xx1 = [tx_1, tx_2, tx_2]
            t_yy1 = [6.75, 6.75, -2] # Would be 7, but doing 6.75 as feature track ratio is 0.75
            t_xx2 = [tx_2, tx_2, tx_1]
            cf = pd.read_csv('../extseq/'+str(tst)+'/'+snn+'_brd_cpg1.csv')
            t_yy2 = [cf['yi'].max(), 0, 0] # ylim
            track.ax.plot(
                t_xx1, t_yy1, 
                c='black', alpha=1, lw=1.5, zorder=2, clip_on=False
            )
            subtrack.ax.plot(
                t_xx2, t_yy2, 
                c='black', alpha=1, lw=1.5, zorder=2, clip_on=False
            )
                            

#block out section
fymi = -10
fyma = 30
bx_st = 43500
bx_en = 51000
# for s in range(len(sregs['nm'])):
# for i in range(len(gv.get_tracks())):
for i in range(6):
    trk = gv.get_tracks()[i]
    trk.ax.fill(
        [bx_st,bx_en,bx_en,bx_st], 
        [fymi,fymi,fyma,fyma], 
        fc='white', alpha=1, zorder=2
    )
    if 'LinkTrack' not in str(gv.get_tracks()[i]):
        if 'TickTrack' not in str(gv.get_tracks()[i]):
            if len(trk.subtracks) > 0:
                strk = trk.subtracks[0]
                strk.ax.fill(
                    [bx_st,bx_en,bx_en,bx_st], 
                    [fymi,fymi,fyma,fyma], 
                    fc='white', alpha=1, zorder=3
                )
    
# for track in gv.get_tracks():
#     track.ax.fill(
#         [40000,50000,50000,40000], 
#         [-20,-20,50,50], 
#         fc="red", alpha=1, zorder=2
#     )
#     subtrack = track.subtracks(sregs['nm'][s])
#
# Track annotations     
# Feature track 0       
tracky0 = gv.get_feature_tracks()[0]
# Track separators
ta_x1 = 44000
ta_x2 = 44150
ta_xt = 44500
ta_xa1 = 0
ta_xa4 = 0
ta_xa5 = 0
ta_xx1 = [ta_x1+ta_xa1, ta_x2+ta_xa1, ta_x2+ta_xa1, ta_x1+ta_xa1]
ta_xx4 = [ta_x1+ta_xa4, ta_x2+ta_xa4, ta_x2+ta_xa4, ta_x1+ta_xa4]
ta_xx5 = [ta_x1+ta_xa5, ta_x2+ta_xa5, ta_x2+ta_xa5, ta_x1+ta_xa5]
ta_ya1 = 0.6
ta_ya4 = 0.9
ta_ya5 = 0.7
ta_ya5corr = 0
ta_ym1 = 0
ta_ym4 = -4
ta_ym5 = -5.85
ta_yy1 = [ta_ym1+ta_ya1, ta_ym1+ta_ya1, ta_ym1-ta_ya1, ta_ym1-ta_ya1]
ta_yy4 = [ta_ym4+ta_ya4, ta_ym4+ta_ya4, ta_ym4-ta_ya4, ta_ym4-ta_ya4]
ta_yy5 = [ta_ym5+ta_ya5, ta_ym5+ta_ya5, ta_ym5-ta_ya5+ta_ya5corr, ta_ym5-ta_ya5+ta_ya5corr]
#
lww = 1.25
lzz = 10
txtsz = 16
#
# Plot separators
tracky0.ax.plot(
    ta_xx1, ta_yy1, 
    c='black', alpha=1, lw=lww, zorder=lzz, clip_on=False
)
tracky0.ax.plot(
    ta_xx4, ta_yy4, 
    c='black', alpha=1, lw=lww, zorder=lzz, clip_on=False
)
tracky0.ax.plot(
    ta_xx5, ta_yy5, 
    c='black', alpha=1, lw=lww, zorder=lzz, clip_on=False
)
# write notes
tracky0.ax.text(
    x=ta_xt, y=ta_ym1, s='Transcript', 
    va='center_baseline', ha='left', size=txtsz
)
tracky0.ax.text(
    x=ta_xt, y=ta_ym4, s='CpG density', 
    va='center_baseline', ha='left', size=txtsz
)
tracky0.ax.text(
    x=ta_xt, y=ta_ym5+(ta_ya5corr/2), s='Alignment', 
    va='center_baseline', ha='left', size=txtsz
)
# 
# Naughty tracks (link ratio 0)
#
ta_xa2 = 0
ta_xx2 = [ta_x1+ta_xa2, ta_x2+ta_xa2, ta_x2+ta_xa2, ta_x1+ta_xa2]
ta_ya2 = 0.8
ta_ym2 = 0
ta_yy2 = [ta_ym2+ta_ya2, ta_ym2+ta_ya2, ta_ym2-ta_ya2, ta_ym2-ta_ya2]
#
ta_xa3 = 0
ta_xx3 = [ta_x1+ta_xa3, ta_x2+ta_xa3, ta_x2+ta_xa3, ta_x1+ta_xa3]
ta_ya3 = 0.8
ta_ym3 = 0
ta_yy3 = [ta_ym3+ta_ya3, ta_ym3+ta_ya3, ta_ym3-ta_ya3, ta_ym3-ta_ya3]
#
tycorr = 0.05 # get text to center
tyadj = 0.2 # drive text apart
#
# Feature track 1
tracky1 = gv.get_feature_tracks()[1]
tracky1.ax.plot(
    ta_xx2, ta_yy2, 
    c='black', alpha=1, lw=lww, zorder=lzz, clip_on=False
)
tracky1.ax.text(
    x=ta_xt, y=ta_ym2+tycorr+tyadj, s='Transposons', 
    va='center_baseline', ha='left', size=txtsz
)
# Feature track 2
tracky2 = gv.get_feature_tracks()[2]
tracky2.ax.plot(
    ta_xx3, ta_yy3, 
    c='black', alpha=1, lw=lww, zorder=lzz, clip_on=False
)
tracky2.ax.text(
    x=ta_xt, y=ta_ym3+tycorr-tyadj, s='Tandem repeats', 
    va='center_baseline', ha='left', size=txtsz
)

#
#
# Custom x axis ticks
# Get max value
mmm = []
for i in range(len(list(gv.get_tracks()))):
    if ':0-' in str(list(gv.get_tracks())[i]):
        mmm.append(gv.get_tracks()[i].end)
# axis labels
lx_st = 0
lx_en = round(max(mmm), -3)
lx_sp = 5000
lx_tk = np.arange(lx_st,lx_en+lx_sp,lx_sp)
lx_lb = [str(int(t/1000)) for t in lx_tk]
plt.xticks(lx_tk, lx_lb)
plt.xlabel('Position (kb)', size=20, labelpad=10)
#
# plt.scatter(45000,1,color='red',s=100000)
#
#
fig.savefig(
    '../pics/syn_reps_foc'+str(tst)+'.png'
)
fig.savefig(
    '../pics/syn_reps_foc'+str(tst)+'.pdf', 
    transparent=True, 
    facecolor=fig.get_facecolor(), #edgecolor='none', 
    bbox_inches='tight', format='pdf'
)
# fig.savefig('../pgv/'+str(tst)+'/'+'pgv_rp.svg')
```
