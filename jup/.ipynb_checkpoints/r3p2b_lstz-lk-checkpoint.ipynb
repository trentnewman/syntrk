```python
import os
import subprocess
from Bio import AlignIO
import pandas as pd
```

```python
tst = 1
wrky = 6
```

```python
dirrie = '../test'+str(tst)+'/'
```

```python
df_r = pd.read_csv('../extseq/'+str(tst)+'/regions.csv')
df_r['len'] = df_r['enn']-df_r['stn']
# df_r.head()
df_rs = df_r[['nm','len']]
lend = df_rs.set_index('nm').to_dict()['len']
lend
```

```python
df_b = pd.read_csv(dirrie+'coords_brd.tsv', sep='\t')
# df_b.head()
```

```python
df_bg = df_b.groupby(['REF_NAME','QUERY_NAME'])['REF_LENGTH','QUERY_LENGTH'].agg(['sum'])
df_bg = df_bg.reset_index(drop=False)
df_bg = df_bg.droplevel(0, axis=1) 
df_bg.columns = ['REF_NAME','QUERY_NAME','REF_LENGTH','QUERY_LENGTH']
df_bg['r_nm'] = [r.split('_')[0] for r in df_bg['REF_NAME']]
df_bg['q_nm'] = [q.split('_')[0] for q in df_bg['QUERY_NAME']]
df_bg['r_ln'] = [lend[r] for r in df_bg['r_nm']]
df_bg['q_ln'] = [lend[q] for q in df_bg['q_nm']]
df_bg['r_pct'] = round((( df_bg['REF_LENGTH'] / df_bg['r_ln'] ) *100),1)
df_bg['q_pct'] = round((( df_bg['QUERY_LENGTH'] / df_bg['q_ln'] ) *100),1)
df_bg
```
