```python
regs = pd.read_csv('../extseq/'+str(tst)+'/regions.csv')
#
estn = []
eenn = []
for i in range(len(regs['nm'])):
    if regs.iloc[i]['dr'] == '+':
        estn.append(regs.iloc[i]['est'] - regs.iloc[i]['st'])
        eenn.append(regs.iloc[i]['een'] - regs.iloc[i]['st'])
    elif regs.iloc[i]['dr'] == '-':
        estn.append(regs.iloc[i]['en'] - regs.iloc[i]['est'])
        eenn.append(regs.iloc[i]['en'] - regs.iloc[i]['een'])
regs['estn'] = estn
regs['eenn'] = eenn
regs.head(2)
```

```python
links = pd.read_csv('../'+tstt+'/coords_brd.tsv', sep='\t')
links = links[links['REF_NAME']!=links['QUERY_NAME']]
links = links.drop(columns=['index'])
links = links.reset_index(drop=True)
links = links.reset_index(drop=False)
links.tail()
```

```python
rrr = []
rs = []
re = []
# ri = []
qqq = []
qs = []
qe = []
# qi = []
for l in range(len(links['index'])):
    #
    ref = links.iloc[l]['REF_NAME']
    ref = ref.split('.fasta')[0]
    ref = ref.replace('_brd','')
    refs = regs[regs['nm']==ref]['st'].iloc[0]
#     refs = refs - refs
    refe = regs[regs['nm']==ref]['en'].iloc[0]
#     refe = refe - refs
    rrr.append(ref)
    rs.append(refs)
    re.append(refe)
    #
    qur = links.iloc[l]['QUERY_NAME']
    qur = qur.split('.fasta')[0]
    qur = qur.replace('_brd','')
    qurs = regs[regs['nm']==qur]['st'].iloc[0]
#     qurs = qurs - qurs
    qure = regs[regs['nm']==qur]['en'].iloc[0]
#     qure = qure - qurs
    qqq.append(qur)
    qs.append(qurs)
    qe.append(qure)
#
links['rrr'] = rrr
links['rs'] = rs
links['re'] = re
links['qqq'] = qqq
links['qs'] = qs
links['qe'] = qe
links.head(2)
#
slinks = links
```

```python
#
df = pd.read_csv('../cmpar/cmpr'+str(tst)+'_brd.csv',sep=',')
cspc = list(df['a'])+[df['b'][len(df['b'])-1]]
# cspc = [s for s in cspc if s != 'asm-MicOch1.0-loc-NC_022015.1:35335047-35512082.fa']
lennie = []
asm = []
cl = []
for s in cspc:
#     if s != 'asm-MicOch1.0-loc-NC_022015.1:35335047-35512082.fa':
    sq = SeqIO.read('../extseq/'+str(tst)+'/'+s, 'fasta')
    lennie.append(len(sq.seq))
    assm = s.split('.fasta')[0]
    assm = assm.replace('_brd','')
#     assm = s.split('-loc')[0].replace('asm-','')
    asm.append(assm)
    cll = regs[regs['nm']==assm]['nm'].iloc[0]
    cl.append(cll)
sll = pd.DataFrame([cspc,asm,cl,lennie]).T
sll.columns = ['cspc','asm','cl','lennie']
sll = sll.reset_index(drop=False)
#
lennie_a = []
for i in range(len(sll['index'])):
    l = sll.iloc[i]['lennie']
    lennie_a.append(l)
#     if sll.iloc[i]['cl'] == 'Striped gopher':
#         lennie_a.append(1000000)
#     else:
#         lennie_a.append(l)
sll['lennie_a'] = lennie_a
sll.head()
```

```python
annie = pd.read_csv('../ann/ann_comb'+str(tst)+'_region.csv', sep='\t')
asn = []
aen = []
for i in range(len(annie['start'])):
    nstt = regs[regs['nm']==annie.iloc[i]['sp']]['st'].iloc[0]
    asn.append(annie.iloc[i]['start_rc']-nstt)
    aen.append(annie.iloc[i]['end_rc']-nstt)
annie['startn'] = asn
annie['endn'] = aen
annie['nm_it'] = '\\it{' + annie['nm'] + '}'
annie.tail()
```

```python
gv = GenomeViz(
    fig_width=10, # at 0.665\textwidth
    feature_track_ratio=1, # keeping constant, 1
    fig_track_height=0.37, # overall height of combined subtracks+links, 0.37
    link_track_ratio=0.5, #0.5
    tick_style="axis",
    align_type='center',
    # track_spines=True,
    tick_labelsize = 17,
)

uss = regs['nm'].unique()
# uss = ['Human','Mouse'] # subset
# uss = ['Human', 'Mouse', 'Elephant', 'Echidna', 'Platypus', 'Agilis'] # subset
# uss = ['Human'] # subset

ssll = sll
ssll = ssll[sll['asm'].isin(uss)] 
for i in range(len(ssll['index'])):
    sssll = ssll.iloc[i]
    track = gv.add_feature_track(
#         name=sglist['ff'], size=sglist['ss']
        name=sssll['cl'], 
        size=sssll['lennie_a'],
        linewidth=0
    )
    track.ratio=0.5 #0.5 # interaction: fig_track_height, link_track_ratio
    
    track.add_subtrack(name=sssll['cl']) # Add subtrack
    san = annie[annie['sp']==sssll['cl']]
    san = san.reset_index(drop=True)
    san = san.reset_index(drop=False)

    # Custom track label position
    track.labelsize=0 # Hide main label
    track.set_sublabel(
        sssll['cl'],
        sublabel_kws=dict(
            alpha=1, y=-1, 
            x=-15000, # position of track label away from track
            va='center', ha='right',
            fontsize=22
        )
    )
    
    for an in range(len(san['index'])):
        sg = san.iloc[an]
        if sg['startn'] >= 0:
            if sg['endn'] < sssll['lennie_a']:
                track.add_feature(
                    sg['startn'], sg['endn'], sg['pm'], 
                    # label=sg['nm_it'], labelsize=20,
                    # labelrotation=0, labelvpos='center', labelha='center',
                    arrow_shaft_ratio=1.0, facecolor="white", linewidth=1,
                    size_ratio=0.75,
                    patch_kws=dict(
                        zorder=1, joinstyle='round')
                )
    
# Add links
sslinks = slinks[ (slinks['rrr'].isin(uss)) & (slinks['qqq'].isin(uss)) ]
for i in range(len(sslinks['index'])):
    sslink = sslinks.iloc[i]
    link_data1 = (sslink['rrr'], sslink['REF_START'], sslink['REF_END'])
    link_data2 = (sslink['qqq'], sslink['QUERY_START'], sslink['QUERY_END'])
    gv.add_link(link_data1, link_data2, 
                normal_color="black", inverted_color="orange", 
                curve=True)  

# Attempts to shorten distance to axis
# for tracky in gv.get_feature_tracks():
#     tracky.ratio
# tracky.ratio = 1
# subtrack = track.get_subtrack()
# subtrack.size = 50
# subtrack.ylim([0,2])
# track.ylim([0,2])
# tracky.size = 50
# tracky.ratio = 50

### plotfig
fig = gv.plotfig()
# Can't access ax property before calling 'plotfig' method
###

# Custom feature label positions
for i in range(len(ssll['index'])):
    sssll = ssll.iloc[i]
    san = annie[annie['sp']==sssll['cl']]
    snn = ssll['cl'][i]
    san = san.reset_index(drop=False)    
    for track in gv.get_feature_tracks():
        if track.name == snn:
            subtrack = track.get_subtrack(snn)
            for an in range(len(san['index'])):
                sg = san.iloc[an]
                if sg['startn'] >= 0:
                    if sg['endn'] < sssll['lennie_a']:
                        track.ax.text(
                            x=subtrack.offset+(sg['startn'] + sg['endn'])/2, 
                            y=-0.15, # accessing this parameter is the purpose of this section
                            size=18, # consider affect with y
                            s=sg['nm_it'], 
                            va='center', ha='center', 
                            c='black', 
                            path_effects=[pe.withStroke(linewidth=1.5, foreground='white')]
                        )

# Plot CpG density
for s in range(len(ssll['cl'])):
    snn = ssll['cl'][s]
    for track in gv.get_feature_tracks():
        if track.name == snn:
            subtrack = track.get_subtrack(ssll['cl'][s])
            x = np.arange(0, track.end, 100)
            y = np.random.randint(0, 100, len(x))
            cf = pd.read_csv('../extseq/'+str(tst)+'/'+snn+'_brd_cpg1.csv')
#             subtrack.ax.fill_between(x, y, color="grey")
            subtrack.ax.fill_between(
#                 cf['xi'], # when left align
                subtrack.offset+cf['xi'], 
                cf['yi'], color="grey")
            subtrack.ax.plot(
                subtrack.offset+cf['xi'], 
                cf['yi'], color="black", linewidth=1)
            symn, symx = 0, 20
            # subtrack.ax.set_ylim(symn, symx)
            subtrack.ax.set_ylim(0,cf['yi'].max())            

ann_regs = [
    ['Human', 200000, 220000],
    ['Mouse', 160000, 285000],
    ['Elephant', 270000, 300000],
    ['Echidna', 320000, 345000],
    ['Platypus', 320000, 345000],
    ['Agilis', 320000, 345000],
    ['Opossum', 320000, 345000],
    ['Koala', 290000, 310000],
    ['Tammar', 290000, 310000],
    ['Brushtail', 320000, 345000],
]
dan_regs = pd.DataFrame(ann_regs)
dan_regs.columns = ['sp', 'st', 'en']
#
# Highlight region of interest
for s in range(len(ssll['cl'])):
    snn = ssll['cl'][s]
    for track in gv.get_feature_tracks():
        if track.name == snn:
            subtrack = track.get_subtrack(ssll['cl'][s])
            dregs = regs
            dregs = dregs[dregs['nm']==ssll['cl'][s]]
            xx = [
                dregs['estn'].iloc[0], dregs['eenn'].iloc[0],
                dregs['eenn'].iloc[0], dregs['estn'].iloc[0]
            ]
            # Region highlights
            r_xx0 = [xi+subtrack.offset for xi in xx]
            r_yy1 = [-1, -1, 1, 1]
            r_yy2 = [symn, symn, symx, symx] # ylim
            track.ax.fill(
                r_xx0, r_yy1, 
                fc="lime", alpha=0.175, zorder=2
            )
            subtrack.ax.fill(
                r_xx0, r_yy2, 
                fc="lime", alpha=0.175, zorder=2
            )
            # Track separators
            tx_1 = -4000
            tx_2 = -6000
            t_xx1 = [tx_1, tx_2, tx_2]
            t_yy1 = [1, 1, -1]
            t_xx2 = [tx_2, tx_2, tx_1]
            cf = pd.read_csv('../extseq/'+str(tst)+'/'+snn+'_brd_cpg1.csv')
            t_yy2 = [cf['yi'].max(), 0, 0] # ylim
            track.ax.plot(
                t_xx1, t_yy1, 
                c='black', alpha=1, lw=1.5, zorder=2, clip_on=False
            )
            subtrack.ax.plot(
                t_xx2, t_yy2, 
                c='black', alpha=1, lw=1.5, zorder=2, clip_on=False
            )
#  
#plt.plot([0,0],[-1,1], c='red', alpha=1, lw=3, zorder=2)

#
# Attempts to shorten distance to axis
# gv.get_tracks()[2].size = 5
# track.size = 5
# track.spines = False
# gv.get_tracks()[2].linewidth = 10
# plt.ylim([0,1])
# for tracky in gv.get_feature_tracks():
#     tracky
# track.ylim([0,2])
#
#
# Custom x axis ticks
# Get max value
mmm = []
for i in range(len(list(gv.get_tracks()))):
    if ':0-' in str(list(gv.get_tracks())[i]):
        mmm.append(gv.get_tracks()[i].end)
# axis labels
lx_st = 0
lx_en = round(max(mmm), -5)
lx_sp = 100000
lx_tk = np.arange(lx_st,lx_en+lx_sp,lx_sp)
lx_lb = [str(int(t/1000)) for t in lx_tk]
plt.xticks(lx_tk, lx_lb)
plt.xlabel('Position (kb)', size=20, labelpad=10)
#
#
plt.rcParams['savefig.facecolor']='None'
# fig.rcParams['savefig.facecolor']='None'
#
fig.savefig(
    '../pics/syn_lstz_brd'+str(tst)+'.png', 
    dpi=300, bbox_inches='tight'
)
fig.savefig(
    '../pics/syn_lstz_brd'+str(tst)+'.pdf', 
    transparent=True, 
    facecolor=fig.get_facecolor(), #edgecolor='none', 
    bbox_inches='tight', format='pdf'
)
# plt.savefig(
#     '../pics/syn_lstz_brd'+str(tst)+'_2.pdf', 
#     transparent=True, 
#     facecolor=fig.get_facecolor(), #edgecolor='none', 
#     bbox_inches='tight', format='pdf'
# )
```
